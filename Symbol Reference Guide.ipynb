## Example Proof Layout in Markdown

**Theorem:** If $n$ is even, then $n^2$ is even.

**Proof:**  
Assume $n$ is even.  
Then $n = 2k$ for some integer $k$.  
Then $n^2 = (2k)^2 = 4k^2 = 2(2k^2)$.  
Since $2k^2$ is an integer, $n^2$ is even.  
$\blacksquare$
