### 1.2.1 Definitions

The derivative of a function $f$ is found by:  

$$
f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}
$$

Essentially we are finding the slope of a tangent line at $x$.

For a simple linear function such as $f(x) = 3x + 4$ the derivate is 3 at all points since it is a straight line.

For more complex functions we can imagine the slope between two points on a graph by plotting a line connecting them as a chord. The slope is the average rate of change between both points and as we reduce the distance between them to 0 we find the instantaneous rate of change at each point.

As $\Delta x$ approaches 0 the secant line becomes equal to the tangent line at $x$.


**Notation and Higher Order Derivatives**

The *second derivative* is the derivative of the derivative and so on for higher order derivatives.  

There are various notations for representing the derivative of $f(x)$

$f'(x)$ and $\frac{df}{dx}$ are common notations for the first derivative

$f''(x)$ and $\frac{d^{2}f}{dx^{2}}$ are common notations for the second derivative   

$f^{m}(x)$ or $\frac{d^{m}(x)}{dx^{m}}$ are analogous to the *m*th derivative  


The *differential operator* $d$  is used to describe dynamics in continuous time.

The *total derivative* of a function of time $t$ or $f(t)$ is given by:
$$df(t) = f'(t)dt$$

Essentially we are looking to find the change in the value of $f(t)$ with respect to a small change in time $t$.  

### 1.2.2 Rules for Differentiation

**Power Rule**

$$\frac{d}{dx}(ax^{n}) = nax^{(n - 1)}$$

**Exponential Rule**
$$\frac{d}{dx}(e^{x}) = e^{x}$$

**Logarithm Rule**
$$\frac{d}{dx}(lnx) = \frac{1}{x}$$

**Chain Rule**
$$\frac{d}{dx}f(g(x)) = f'(g(x)) \cdot g'(x)$$

**Sum Rule**
$$\frac{d}{dx}(f(x) + g(x)) = f'(x) + g'(x)$$

**Product Rule**
$$\frac{d}{dx}(f(x) \cdot g(x)) = f'(x) \cdot g(x) + g'(x) \cdot f(x)$$

**Quotient Rule**
$$\frac{d}{dx}\left(\frac{g(x)}{f(x)}\right) = \frac{f'(x) \cdot g(x) - g'(x) \cdot f(x)}{f(x)^{2}}$$

**Constants**
$$\frac{d}{dx}c = 0$$

So in our interactive example $f(x) = 2x^{3}$ we simply have to apply the power rule to analytically differentiate.  

The first derivative to find $f'(x)$ in our example $3 \cdot 2x^{(3 - 1)}$ or $6x^{2}$ and this is represented by tangent indicated by the red dashed line on our graph.  

So the slope or instantaneous rate of change $\frac{df}{dx}$ or $f'(x)$ anywhere on our function will be our $x$ value squared then multiplied by $6$.  

The second derivative $\frac{d^{2}f}{dx^{2}}$ or  $f''(x)$ would then be $12x$.

### 1.2.3 Monotonic, Concave and Convex Functions

**Monotonic Increasing & Monotonic Decreasing**  

A differentiable function is *strictly monotonic increasing* if $f'(x) > 0 \ \text{for all} \ x$.  
It is *strictly monotonic decreasing* if $f'(x) < 0 \ \text{for all} \ x$.  
- Sometimes the '*strictly*' is dropped in cases we allow the function to be flat i.e. having a zero derivative for some $x$.


**Concave & Convex**  

If the second derivative $f''(x) \lt 0$ than $f$ is *strictly concave*.
If $f''(x) \gt 0$ it is *strictly convex*.

The '*strictly*' is dropped in cases that are not strict inequalities i.e. $f''(x) \le 0 \ \text{or} \ f''(x) \ge 0$.

**Inflection Points**

Some functions may be concave on some ranges of $x$ and convex on other ranges of $x$.  
These are considered *inflection points*.  

The function $f(x) = 2x^3$ used in the example above clearly demonstrates this behaviour. It is clear to see an inflection point where the graph changes from being concave for negative values of $x$ to convex for positive values of $x$.  

$f''(x) = 12x$ is the second derivative of the function and $12x > 0$ when for positive values of $x$ and $12x < 0$ for negative  values of $x$. Setting $12x = 0$ shows the inflection point happens when $x = 0$.

$\text{Formal Definitions}$

$f$ is *strictly concave* if and only if
$$f(px_1 + (1 - p)x_2) \gt pf(x_1) + (1 - p) \cdot f(x_2) \ \text{for every} \ p \in [0,1].$$  

$f$ is *strictly convex* if and only if
$$f(px_1 + (1 - p)x_2) \lt pf(x_1) + (1 - p) \cdot f(x_2) \ \text{for every} \ p \in [0,1].$$  

<br>  

That is to say for any two points in the range $[x_1, x_2]$ will lie above the chord connecting $f(x_1) \ \text{and} \ f(x_2)$ if it is concave and below if it is convex.

$\text{Examples in Finance}$

**Price of a bond** - a convex monotonic decreasing function of its yield, hence the second derivative of a bond's price with respect to its yield is known as its *convexity*

### 1.2.4 Stationary Points and Optimization

**Stationary Points**

When $f'(x) = 0$ it is a *stationary point*.  
The tangent will be a horizontal line with a slope of 0.  

We can use the values of our stationary points to determine if they represent a *local minimum*, *local maximum*, or a *saddle point*.

When $f''(x) > 0$ at the stationary point $f$ has a *local minimum*.  
When $f''(x) < 0$ at the stationary point $f$ has a *local maximum*.  
If $f'(x) = 0$ and $f''(x) = 0$ then the stationary point should be evaluated further.  
- Find the value of $f''(x)$ on both sides of $x$ and evaluate $f''(x - \epsilon)$ and $f''(x + \epsilon)$
    - If both are > 0 *local minimum*
    - If both are < 0 *local maximum*
    - If opposite signs <> *saddle point*


Finding these critical points, inflection points, minimum and maximum values are crucial for optimization problems to come.

### 1.2.5 Integration

So far we've focused on simple ordinary derivatives that examine how something changes in response to slight changes to a variable.  

 The integration process is the opposite, but helps us add up all these small changes. 

 $\text{Example of a Definite Integral}$

 Lets define a function as $f(x) = x^{2} + 2x$

 The derivative shows us that the rate $f(x)$ is changing with respect to $x$ is
 $$f'(x) = 2x + 2$$

 But to see the *total change* between any two points I can find the *definite integral*.

 $$\int_{a}^{b} \left( x^2 + 2x \right) \ dx \ = F(b) - F(a)$$

 Where $a$ is the lower limit and $b$ is the upper limit for a range of values I want to calculate the area.

First we integrate $f(x)$ by the opposite process of differentiation to find $F(x)$ and solve for $F(b) - F(a)$
 
$$F(x) = \frac{x^{3}}{3} + x^{2}$$

$$= \int_{2}^{4} \left( x^2 + 2x \right) \, dx = \left[ \frac{x^3}{3} + x^2 \right]_2^4$$

$$= \left( \frac{4^3}{3} + 4^2 \right) - \left( \frac{2^3}{3} + 2^2 \right)$$

$$\approx 30.66667$$
